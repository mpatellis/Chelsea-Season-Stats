```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("dataframe.csv", "r")
df = pd.read_csv(f)
```

```python
df.head()
```

```python
apperances = df['appearences'] > 5
goals = df['goals'] > 0
assists = df['assists'] > 1
df = df[apperances & goals & assists]
```

```python
mingoal = pd.Series(df['minutes']/df['goals'])
minassist = pd.Series(df['minutes']/df['assists'])
df['min_goal'] = mingoal
df['min_assist'] = minassist
goalmin = df['min_goal'] < 600
assistmin = df['min_assist'] < 600
df = df[assistmin & goalmin]
```

```python
plt.style.use('fivethirtyeight')
mingoal = df.min_goal
minassist = df.min_assist
name = df.name
```

```python
plt.figure(dpi=500)
plt.scatter(mingoal, minassist, c='#034694')
plt.axis('square')
plt.title('Minutes per goal vs minutes per assist')
plt.xlabel('Minutes per goal')
plt.ylabel('Minutes per assist')
i = 0
while i < len(name):
    plt.annotate(name.iloc[i], (mingoal.iloc[i], minassist.iloc[i]))
    i = i +1
```
